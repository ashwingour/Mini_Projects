```python
import os
def rename_files():
    #Getting files name form a specific folder
    file_list= os.listdir(r"C:\prank");
    #print(file_list);
    saved_path = os.getcwd()
    print("Current Working Directory is  "+saved_path)
    os.chdir(r"C:\prank")
    saved_path_1 = os.getcwd()
    print("Current Working Directory is  "+saved_path_1)

    
    #renaming each and  every file
    for file_name in file_list:
        print("Old Name - "+file_name)
        table = str.maketrans(dict.fromkeys('0123456789'))
        new_name = file_name.translate(table)
        os.rename(file_name, new_name )
        print("New Name - "+new_name)
    #os.chdir(saved_path)
    
rename_files()
```
